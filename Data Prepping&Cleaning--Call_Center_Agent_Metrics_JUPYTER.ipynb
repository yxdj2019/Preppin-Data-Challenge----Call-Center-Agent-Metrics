```python
df=pd.ExcelFile('MetricData2021.xlsx')
```

```python
# read all the sheets
df_Jan=df.parse(0)
df_Feb=df.parse(1)
df_Mar=df.parse(2)
df_Apr=df.parse(3)
df_May=df.parse(4)
df_Jun=df.parse(5)
df_Jul=df.parse(6)
df_Aug=df.parse(7)
df_Sep=df.parse(8)
df_Oct=df.parse(9)
df_Nov=df.parse(10)
df_Dec=df.parse(11)
df_Jan.head()
```

```python
# insert the AgentID that do not exist in sheets: make sure each sheet has all the agentID
s=[df_Jan, df_Feb, df_Mar,df_Apr, df_May,df_Jun ,df_Jul, df_Aug,df_Sep,df_Oct, df_Nov,df_Dec]
a=np.arange(135)+1
for na in range(len(s)):
    b=s[na]['AgentID'].values
    c=np.setdiff1d(a,b)  #Find the set difference of two arrays. Return the unique values in ar1 that are not in ar2.
    s[na]=pd.concat([s[na],pd.DataFrame(c,columns=['AgentID'])],ignore_index=True, sort=False)
[df_Jan, df_Feb, df_Mar,df_Apr, df_May,df_Jun ,df_Jul, df_Aug,df_Sep,df_Oct, df_Nov,df_Dec]=s
```

```python
# add Month start date for each sheet
df_Jan['Month Start Date']='2021-01-01'
df_Feb['Month Start Date']='2021-02-01'
df_Mar['Month Start Date']='2021-03-01'
df_Apr['Month Start Date']='2021-04-01'
df_May['Month Start Date']='2021-05-01'
df_Jun['Month Start Date']='2021-06-01'
df_Jul['Month Start Date']='2021-07-01'
df_Aug['Month Start Date']='2021-08-01'
df_Sep['Month Start Date']='2021-09-01'
df_Oct['Month Start Date']='2021-10-01'
df_Nov['Month Start Date']='2021-11-01'
df_Dec['Month Start Date']='2021-12-01'
```

```python
# union all the sheets
s=[df_Jan, df_Feb, df_Mar,df_Apr, df_May,df_Jun ,df_Jul, df_Aug,df_Sep,df_Oct, df_Nov,df_Dec]
df_Months=pd.concat(s, ignore_index=True, sort=False)
df_Months.shape
```

```python
# find indexes for cleaning
index=df_Months.loc[:,['Calls Answered','Answered']].isna().sum(axis=1)==1
index
```

```python
# cleaning df_Months : Combine--column Answered and Calls Answered
df_Months.loc[index,'Calls Answered'] = df_Months.loc[index,'Answered'].fillna(0)+df_Months.loc[index, 'Calls Answered'].fillna(0)
```

```python
# drop Answered column
df_Months=df_Months.drop(columns=['Answered'])
```

```python
# cleaning df_Months : Combine--column Not Answered and Calls Not Answered
df_Months.loc[index,'Calls Not Answered'] = df_Months.loc[index,'Not Answered'].fillna(0)+df_Months.loc[index,'Calls Not Answered'].fillna(0)
```

```python
# drop Not Answered column
df_Months=df_Months.drop(columns=['Not Answered'])
```

```python
df_Months.head()
```

```python
# cleaning df_Months : Combine--column Calls Offered and Offered
df_Months.loc[index,'Calls Offered'] = df_Months.loc[index,'Offered'].fillna(0)+df_Months.loc[index,'Calls Offered'].fillna(0)
```

```python
# drop Offered column
df_Months=df_Months.drop(columns=['Offered'])
```

```python
# calculate Not Answered Rate
df_Months['Not Answered Rate']=df_Months['Calls Not Answered']/df_Months['Calls Offered']
df_Months.head()
```

```python
# Met Not Answered Rate
df_Months['Met Not Answered Rate?']=df_Months['Not Answered Rate']<0.05
df_Months.head()
```

```python
# read PeopleData.xlsx
df=pd.ExcelFile('PeopleData.xlsx')
```

```python
len(df.sheet_names)
```

```python
df_People=df.parse(0)
df_Leaders=df.parse(1)
df.Location=df.parse(2)
df_Goals=df.parse(3)
```

```python
# join People and Leaders
df_temp=df_People.merge(df_Leaders, left_on='Leader 1', right_on='id', how='left')
```

```python
# sort values by id
df_temp=df_temp.sort_values(by='id_x')
```

```python
# drop leader id
df_temp=df_temp.drop(columns=['id_y'])
```

```python
df_temp
```

```python
# combine first name and last name for agent
# pandas.Series.str.cat
df_temp['Agent Name']=df_temp['first_name_x'].str.cat(df_temp['last_name_x'],join='left', sep=', ', na_rep='-')
```

```python
# combine first name and last name for leader
# pandas.Series.str.cat
df_temp['Leader Name']=df_temp['first_name_y'].str.cat(df_temp['last_name_y'],join='left', sep=', ', na_rep='-')
```

```python
df_temp
```

```python
# Drop first and last name columns
df_temp=df_temp.drop(columns=['first_name_x', 'last_name_x', 'first_name_y', 'last_name_y'])
```

```python
# change id_x to id
df_temp.rename(columns={'id_x':'id'}, inplace=True)
```

```python
# join df_temp and Location
df_pll=df_temp.merge(df.Location, on='Location ID', how='left')
```

```python
# drop location id
df_pll=df_pll.drop(columns='Location ID')
```

```python
# Join df_Months and df_pll
df_final=df_pll.merge(df_Months, how='left', right_on='AgentID', left_on='id')
```

```python
# drop AgentID
df_final.drop(columns='AgentID',inplace=True)
```
